```python
import pandas as pd
```

```python
pa_data = pd.read_csv('raw_data/pa.csv')
pa_data
```

```python
ventiladores = pa_data[pa_data['registro.Descrição do bem ou serviço'].str.contains("Respirador")]
ventiladores
```

```python
pa = pd.DataFrame()
pa['nome'] = ventiladores['registro.Descrição do bem ou serviço']
pa['preco'] = ventiladores['registro.Valor unitário']
pa['estado'] = 'PA'
pa['data'] = ventiladores['registro.Data de celebração do contrato']
pa['quantidade'] = ventiladores['registro.Qtde']

pa.to_csv('data/pa.csv', index=False)
```
