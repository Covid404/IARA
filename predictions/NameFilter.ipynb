```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_colwidth', -1)
df = pd.read_csv("predictions.csv")
df.iloc[69]
```

```python
def teste(nome):
    nome = nome.split()
    #print ([word for word in nome])
    nome_final = ' '.join(
        [
            word.upper() for word in nome 
                 if word.replace('(', '').replace(')', '').replace(',', '').replace('.', '').lower()
                        in ['ventilador', 'artificial', 'artificiais', 
                            'pulmonar', 'mecanico', 'microprocessado', 'mecânico', 
                            'ventiladores', 'pulmonares', 'respiradores'
                           ]
        ])

    return nome_final

#teste(df['nome'][0])
#df['nome'] = [word for word in df['nome'].str if word in ['VENTILADOR']]
df_teste = df.copy()
df_teste['nome'] = df['nome'].apply(teste)
```
